```python
import pandas as pd
from astropy.io import ascii
import matplotlib.pyplot as plt
```

```python
# Load Dataset
df = ascii.read('../data/raw/robovetter/inj/kplr_dr25_inj1_robovetter_input.txt').to_pandas()
df.head()
```

### Describe and Plot Distribution

```python
for n in list(df.columns.values):
    if n in ['Data_Set', 'TCE_ID']: continue
    print(n)
    print(df[n].describe())
    df[n].hist()
    plt.show()
    print('='*80)
```
